# Define Dataset (20 Points)

```python
import glob
from PIL import Image as im

convert_tensor = transforms.ToTensor()
file_list = [convert_tensor(im.open(path).convert('L'))[0] for path in glob.glob("/content/efigi-1.6/png/*.png")]
labels = df['class_label'].values.tolist()

X_train, X_dev, y_train, y_dev = train_test_split(file_list, labels, test_size=0.4)
X_dev, X_test, y_dev, y_test = train_test_split(X_dev, y_dev, test_size=0.5)
```

```python
###############################################################
##        Write your dataset class for loading images        ##
##                        Your Code                          ##
###############################################################
class GalaxyDataSet(Dataset):
    def __init__(self, file_list, labels):
        self.file_list = file_list
        self.labels = labels
        
    def __len__(self):
        return len(self.labels)

    def __getitem__(self, index):
        image = self.file_list[index]
        label = self.labels[index]
        return image, label
```

```python
batch_size = 128

train_loader = DataLoader(GalaxyDataSet(X_train, y_train), batch_size=batch_size)
dev_loader = DataLoader(GalaxyDataSet(X_dev, y_dev), batch_size=batch_size)
test_loader = DataLoader(GalaxyDataSet(X_test, y_test), batch_size=batch_size)
```

# Define Model (20 Points)

```python
#####################################
##        Define your model        ##
##            Your Code            ##
#####################################
class MLP(nn.Module):
    def __init__(self, d_layer):
        super().__init__()
        self.d_layer = d_layer
        layer_list = [nn.Linear(d_layer[l], d_layer[l+1]) for l in range(len(d_layer) - 1)]
        self.linears = nn.ModuleList(layer_list)

    def forward(self, X):
        X = X.view(-1, self.d_layer[0])
        for layer in self.linears[:-1]:
            X = F.relu(layer(X))
        return F.softmax(self.linears[-1](X), dim=1)
```

# Train Model (30 Points)

```python
def train_val_model(model, criterion, optimizer, dataloaders, num_epochs):
    best_acc = 0.0
    losses, accuracies = dict(train=[], val=[]), dict(train=[], val=[])

    for epoch in range(num_epochs):
        print('Epoch {}/{}'.format(epoch, num_epochs - 1))
        print('-' * 10)

        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()  # Set model to training mode
            else:
                model.eval()   # Set model to evaluate mode

            running_loss = 0.0
            running_corrects = 0
            nsamples = 0

            for inputs, labels in dataloaders[phase]:
                optimizer.zero_grad()

                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                nsamples += inputs.shape[0]
                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

            epoch_loss = running_loss / nsamples
            epoch_acc = running_corrects.double() / nsamples

            losses[phase].append(epoch_loss)
            accuracies[phase].append(epoch_acc)

            print('{} Loss: {:.4f} Acc: {:.2f}%'.format(phase, epoch_loss, 100 * epoch_acc))

            if phase == 'val' and epoch_acc > best_acc:
                best_acc = epoch_acc

    print('Best val Acc: {:.2f}%'.format(100 * best_acc))
    return model, losses, accuracies
```

```python
######################################################################
##        Instantiate model, define hyper parameters, optimizer,    ##
##        loss function ant etc                                     ##
######################################################################
INPUT_DIM = 255 * 255
OUTPUT_DIM = 4
model = MLP([INPUT_DIM, 512, 256, 128, 64, OUTPUT_DIM])
model = MLP([INPUT_DIM, 64, OUTPUT_DIM])
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()
dataloaders = {"train": train_loader, "val": dev_loader}
num_epochs = 10
model, losses, accuracies = train_val_model(model, criterion, optimizer, dataloaders, num_epochs)
```

```python
##############################################################
##          Plot metrics graph for different epochs         ##
##                        Your Code                         ##
##############################################################
plt.plot(losses['train'], '-b', losses['val'], '--r')
plt.show()
```
