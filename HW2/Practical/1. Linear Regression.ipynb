First we load the required packages

```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
```

Then we load the dataset. Here we chose california housing dataset which contains the price of the houses in california
based on some parameters like its age, count rooms, size and other parameters.

```python
data=fetch_california_housing()
df=pd.DataFrame(data.data,columns=data.feature_names)
target_name=data.target_names[0]
df[target_name]=data.target
df
```

We check if it has any na values in its columns

```python
df.isna().any()
```

Now using StandaradScaler we scale all the features in the data set so they would have zero mean and standard deviation one.

```python
Y=df[target_name]
X=df.drop(columns=[target_name])
X_scaled=pd.DataFrame(StandardScaler().fit_transform(X),columns=X.columns)
print(X_scaled)
print('Mean:')
print(np.mean(X_scaled,axis=0))
print('STD:')
print(np.std(X_scaled,axis=0))
```

Then we split the data into train and test data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_scaled,Y)
X_train
```

We train the data on linear regression model and we compute the mean squared error of it

```python
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,Y_train)
prediction=model.predict(X_test)
mean_squared_error(prediction,Y_test)
```

Now using Lasso regression we try to train the data and we find the best hyper parameter for this model

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
from warnings import simplefilter
from sklearn.exceptions import ConvergenceWarning
simplefilter("ignore", category=ConvergenceWarning)
x=np.arange(0,0.02,0.001)
losses=[]
for i in range(20):
    Lreg=linear_model.Lasso(alpha=i/1000)
    Lreg.fit(X_train,Y_train)
    prediction=Lreg.predict(X_test)
    mean_squared_error(prediction,Y_test)
    losses.append(mean_squared_error(prediction,Y_test))
plt.plot(x,losses)
plt.title('Cost function vs hyperparameter in Lasso Regression')
print(f'The best hyper parameter for Lasso regression is {x[np.argmin(losses)]}')
print(f'The least cost is {min(losses)}')
```

We do the same for Ridge Regression this time:

```python
x=np.arange(50,200,1)
losses=[]
for i in range(150):
    Rreg=linear_model.Ridge(alpha=i+50)
    Rreg.fit(X_train,Y_train)
    prediction=Rreg.predict(X_test)
    mean_squared_error(prediction,Y_test)
    losses.append(mean_squared_error(prediction,Y_test))
plt.plot(x,losses)
plt.title('Cost function vs hyperparameter in Ridge Regression')
print(f'The best hyperparameter for Ridge regression is {x[np.argmin(losses)]}')
print(f'The least cost is {min(losses)}')
```

```python
from sklearn.linear_model import ElasticNet
from warnings import simplefilter
from sklearn.exceptions import ConvergenceWarning
simplefilter("ignore", category=ConvergenceWarning)
alpha=np.arange(0.005,0.015,0.001)
l1_ratio=np.arange(0.015,0.025,0.001)
losses=[]
for i in alpha:
    for j in l1_ratio:
        elastic_net = ElasticNet(alpha = i, l1_ratio = j)
        elastic_net.fit(X_train,Y_train)
        prediction=elastic_net.predict(X_test)
        losses.append(mean_squared_error(prediction,Y_test))
print(f'The best hyperparameters for Elastic net regression is alpha={alpha[np.argmin(losses)//10]}, l1_ratio={l1_ratio[np.argmin(losses)%10]}')
print(f'The least cost is {min(losses)}')
```

Elasticnet regression is a combination of two Lasso and Ridge Regression and somehow overcomes these two models weakness.
